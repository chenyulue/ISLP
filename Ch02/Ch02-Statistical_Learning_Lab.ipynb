To create a matrix of numbers, we can also use `np.array`. As the help information above shows, `np.array` can create multi-dimensional arrays, and a matrix is a 2-dimensional array.

The `np.sqrt` function returns the square root of each element of a vector or matrix. The expression `x**2` or `np.power(x, 2)` raises each element of `x` to the power 2; fractional and negative powers are allowed, but negative integer powers are not.

```python
np.sqrt(x)
```

The `np.random.normal()` function generates an array from the "standard normal" distribution. To create a vector with n elements, just give the option `size=n` to the function. Each time we call this function, we will get a different vector. Here we create two correlated sets of numbers, `x` and `y`, and use the `np.correlate` function to compute the correlation between them.

```python
x = np.random.normal(size=50)
y = x + np.random.normal(50, 0.1, size=50)
np.correlate(x, y)
```

By default, `np.random.normal()` creates a standard normal random varialbes with a mean 0f 0 and a standard diviation of 1. However, the mean and standard deviation can be altered by passing the first (the mean) and the second (the std) arguments, as illustrated above. We can also use `np.seed()` function to make our code to reproduce the exact same set of random numbers

```python
np.random.seed(1303)
np.random.normal(size=50)
```

The `np.mean()` and `np.var()` functions can be used to computed the mean and variance of a vector of numbers. Applying `np.sqrt()` to the output of `np.var()` will give the standard deviation. Or simply use the `np.std()` function.

```python
np.random.seed(3)
y = np.random.normal(size=100)
np.mean(y)
```

```python
np.var(y)
```

```python
np.sqrt(np.var(y))
```

```python
np.std(y)
```

## 2.3.2 Graphics
In python, we usually use the `matplotlib` package to plot data. The recommended way to use it for a beginner is `import matplotlib.pyplot as plt`, and plot the data by the function `plt.plot(x, y)` to produce a lineplot ot the numbers in `x` versus the numbers in `y`.

```python
import matplotlib.pyplot as plt
x = np.random.normal(size=100)
y = np.random.normal(size=100)
plt.plot(x, y)
```

There are also many options that can be passed in to the `plt.plot()` function, such as `marker='o'` together with `ls=''` to produce a scatterplot. Alternatively, it is recommended to use `plt.scatter` to obtain a scatterplot. 

```python
plt.plot(x, y, marker='o', ls='')
```

```python
plt.scatter(x, y)
```

To set the x-label, y-label, and the plot title, we can use `plt.ylabel()`, `plt.xlabel()` and `plt.title()` to set the corresponding options:

```python
plt.plot(x, y, marker='o', ls='')
plt.xlabel('this is the x-axis')
plt.ylabel('this is the y-axis')
plt.title('Plot of X vs Y')
plt.savefig('Figure.pdf')
```

To save the output of a python plot, we can use the function `plt.savefig()`. The file type that we would like to create depends on the file name that we pass in to the function. To create a pdf, the function takes `Figure.pdf` as the first parameter, and to create a jpeg, `Figure.jpg` is the first parametr.

The function `np.arange` and `np.linspace` can be used to create a sequence of numbers.
* `np.arange(a, b)` creates a vector of integers between `a` and `b`(excluded)
* `np.linspace(0, 1, 10)` makes a sequence of 10 numbers that are equally spaced between 0 and 1

```python
x = np.arange(1, 10)
x
```

```python
x = np.linspace(-np.pi, np.pi, 50)
x
```

In python, we can use `plt.contour` to create a *contour plot* in order to represent three-dimensional data. The function basically takes three arguments:
* X: a vector of the `x` value (the first dimension, with the length M equal to the number of columns of `Z`)
* Y: a vector of the `y` value (the second dimension, with the length N equal to the number of rows of `Z`)
* Z: a (N, M) matrix whose elements correspond to the `z` value (the third dimension) for each pair of `(x,y)` coordinates.
There are also many other options to fine-tune the output of the `plt.contour` function. You can type `?plt.contour` to learn more.

```python
y = x
X, Y = np.meshgrid(x, y)
z = np.cos(Y) / (1 + x**2)
plt.contour(x, y, z)
```

```python
plt.contour(x, y, z, levels=45)
```

```python
za = (z - np.transpose(z)) / 2
plt.contour(x, y, za, levels=15)
```

In python, `plt.imshow` works the same way as R's `image()`, except that the x-axis and y-axis have to be specified manually.

```python
plt.imshow(za, extent=[-np.pi, np.pi, -np.pi, np.pi], cmap='hot')
```

It's a little complicated to produce a three-dimensional plot using `matplotlib`. We can use `view_init(elev, azim)` to set the view angle, where `elev` is the elevation angle in the z plane, and `azim` is the azimuth angle in the x,y plane.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, za, cmap='coolwarm')
ax.view_init(elev=10, azim=10)
```

## 2.3.3 Indexing Data

```python
A = np.arange(1,17).reshape(4,4, order='F')
A
```

```python
A[1,2]
```

In python, indexing data is zero-indexed, which means `A[1,2]` selects the element corresponding to the second row (1+1) and the third column (2+1). To select multiple rows and columns at a time, provide `np.ix_` or slices as the indices:

```python
A[np.ix_([0,2],[1,3])]
```

```python
A[0:3, 1:4]
```

```python
A[:,0:2]
```

```python
A[0,:]
```

To include all columns or all rows, use `:` in the corresponding position, as shown above.

An array's `shape` attribute gives a tuple with the number of rows as the first element and the number of columns as the second element:

```python
A.shape
```

## 2.3.4 Loading Data
In python, there are many ways to load data. For data analysis, **pandas** is the most popular package used to manipulate data including loading data. The most import data type in pandas is `DataFrame`, which is similar to *data frame* in R.

```python
import pandas as pd
```

```python
Auto = pd.read_table('../data/Auto.data', sep='\s+')
Auto
```

Since `pd.read_table` can infer the header, you don't need to specify the header manually though you can if you want to. But the default separator used by `pd.read_table` is `\t`, and in the `Auto.data`, the separators include `\t` and space(s), so we need to specify the separator by the option `sep`. Besides, to deal with the missing values, the option `na_values` can be used.

```python
Auto = pd.read_table('../data/Auto.data', sep='\s+', na_values='?')
Auto.loc[32]
```

We can also use `pd.read_csv` to load a csv files.

```python
Auto = pd.read_csv('../data/Auto.csv', na_values='?')
Auto
```

The `DataFrame`'s attribute `shape` gives the numbers of rows and columns, that is, the dimension of the data. We can also remove the rows which contain the missing values by the method `DataFrame.dropna()`.

```python
Auto.shape
```
